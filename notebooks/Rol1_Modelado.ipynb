# Caso SUNBURST

SUNBURST fue un ciberataque de tipo supply chain attack que comprometió versiones del software Orion Platform desarrollado por SolarWinds.

El ataque permitió insertar código malicioso en actualizaciones legítimas del software, afectando múltiples organizaciones gubernamentales y privadas alrededor del mundo.

Este caso representa un escenario adecuado para aplicar conceptos de modelado, integración y calidad de datos.

# Modelo Entidad-Relación

El modelo de datos desarrollado se compone de tres entidades principales:

## Clientes
Representa las organizaciones que utilizan Orion Platform.

## Versiones_Software
Representa las versiones del software y si contienen la vulnerabilidad SUNBURST.

## Instalaciones
Representa las instalaciones realizadas por cada cliente.

### Relaciones
- Un cliente puede tener múltiples instalaciones.
- Una versión de software puede estar presente en múltiples instalaciones.

```python
import pandas as pd
import numpy as np
from faker import Faker
import random
import os
```

```python
fake = Faker()

random.seed(42)
np.random.seed(42)
```

# Generación de Datos Sintéticos - Clientes

Se generan 50 registros relacionados con organizaciones potencialmente afectadas por el incidente SUNBURST.

```python
def generar_cliente(cliente_id):

    tipos_org = [
        "Gobierno",
        "Empresa Privada",
        "Proveedor TI",
        "Entidad Financiera",
        "Universidad"
    ]

    sectores = [
        "Tecnología",
        "Finanzas",
        "Salud",
        "Educación",
        "Defensa",
        "Telecomunicaciones"
    ]

    criticidades = [
        "Alta",
        "Media",
        "Baja"
    ]

    return {

        "cliente_id": cliente_id,

        "nombre_organizacion": fake.company(),

        "tipo_org": random.choice(tipos_org),

        "pais": fake.country(),

        "sector": random.choice(sectores),

        "criticidad": random.choice(criticidades)
    }
```

```python
clientes_lista = []

for i in range(1, 51):

    clientes_lista.append(
        generar_cliente(i)
    )

clientes = pd.DataFrame(clientes_lista)

clientes.head()
```

# Generación de Datos Sintéticos - Instalaciones

Se generan instalaciones asociadas a clientes y versiones de software.

```python
def generar_instalacion(instalacion_id):

    niveles_datos = [
        "Alto",
        "Medio",
        "Bajo"
    ]

    return {

        "instalacion_id": instalacion_id,

        "cliente_id": random.randint(1, 50),

        "version_id": random.randint(1, 8),

        "fecha_instalacion": fake.date_between(
            start_date='-3y',
            end_date='today'
        ),

        "nivel_datos_sensibles": random.choice(
            niveles_datos
        )
    }
```

```python
instalaciones_lista = []

for i in range(1, 101):

    instalaciones_lista.append(
        generar_instalacion(i)
    )

instalaciones = pd.DataFrame(
    instalaciones_lista
)

instalaciones.head()
```

# Exportación de Archivos CSV

Los DataFrames generados se exportan a archivos CSV para permitir su reutilización y análisis posterior.

```python
ruta_actual = os.getcwd()

clientes.to_csv(
    os.path.join(ruta_actual, "clientes.csv"),
    index=False
)

versiones_software.to_csv(
    os.path.join(ruta_actual, "versiones_software.csv"),
    index=False
)

instalaciones.to_csv(
    os.path.join(ruta_actual, "instalaciones.csv"),
    index=False
)

print("Archivos CSV exportados correctamente.")
```
