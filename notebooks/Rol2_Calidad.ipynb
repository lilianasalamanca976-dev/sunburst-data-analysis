## Proyecto: Análisis del incidente SUNBURST (SolarWinds)

Este notebook presenta el análisis de calidad de datos, validación, detección de anomalías y visualización de datos sintéticos generados para simular el incidente cibernético SUNBURST.

Se aplican conceptos del marco DAMA-DMBOK, incluyendo métricas de calidad de datos, integridad referencial y análisis de patrones.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
clientes = pd.read_csv("clientes.csv")
versiones_software = pd.read_csv("versiones_software.csv")
instalaciones = pd.read_csv("instalaciones.csv")
eventos_seguridad = pd.read_csv("eventos_seguridad.csv")
reporte_calidad = pd.read_csv("reporte_calidad.csv")
```

## Exploración inicial de los datos

Se cargan los datasets generados en el Rol 1 y Rol 2 para verificar su estructura y contenido.

```python
clientes.head()
versiones_software.head()
instalaciones.head()
eventos_seguridad.head()
```

## Métricas de calidad de datos

Se evaluaron tres métricas principales basadas en el marco DAMA-DMBOK:

- Completitud
- Exactitud
- Consistencia

```python
reporte_calidad
```

### Análisis

Se observa que la mayoría de las métricas cumplen con los umbrales definidos, excepto la consistencia en la tabla de versiones de software.

Esto indica la presencia de inconsistencias en las fechas, lo cual representa un problema en el ciclo de vida del software.

## Detección de anomalías

Se identificaron patrones anómalos en los datos, incluyendo:

- Eventos de seguridad sospechosos
- Distribución irregular de instalaciones
- Eventos de alta severidad

```python
eventos_seguridad["es_anomalo"].value_counts()
```

```python
sns.countplot(data=eventos_seguridad, x="severidad")
plt.title("Distribución de Severidad de Eventos")
plt.show()
```

```python
sns.countplot(data=eventos_seguridad, x="es_anomalo")
plt.title("Eventos Anómalos vs Normales")
plt.show()
```

```python
sns.countplot(data=eventos_seguridad, x="tipo_evento")
plt.xticks(rotation=45)
plt.title("Tipos de Evento")
plt.show()
```

```python
conteo = instalaciones.groupby("cliente_id").size()

sns.histplot(conteo, bins=10)
plt.title("Instalaciones por Cliente")
plt.show()
```

```python
sns.countplot(data=instalaciones, x="estado_seguridad")
plt.title("Estado de Seguridad")
plt.show()
```

```python
eventos_seguridad["timestamp"] = pd.to_datetime(eventos_seguridad["timestamp"])

eventos_por_fecha = eventos_seguridad.groupby(
    eventos_seguridad["timestamp"].dt.date
).size()

eventos_por_fecha.plot()
plt.title("Eventos en el tiempo")
plt.show()
```

## Simulación del incidente SUNBURST

Se simuló un escenario en el cual múltiples instalaciones estaban potencialmente afectadas por una versión comprometida del software.

Sin embargo, solo un subconjunto reducido fue considerado realmente comprometido, reflejando la naturaleza selectiva del ataque.

Esto permite diferenciar entre exposición al riesgo y compromiso real.
